Zadanie 2
Regular expression
Wstaw zamiast "____" poprawny numer zamówienia zgodny z wyrażeniem regularnym. Poprawny tekst zwróci wynik ewaliuacji 1.

```python
from langchain_classic.evaluation import RegexMatchStringEvaluator

evaluator = RegexMatchStringEvaluator()
result = evaluator.evaluate_strings(
    prediction="____", # <- tu wstaw numer zamówienia
    reference=r"^Order ID: [A-Z]{3}-\d{4}$",
)
print(result['score'])
```

Zadanie 3
Word limit
Uruchom kod. Nastepnie zmniejsz wartości zmienych limit i lower_limit o 5 i uruchom ponownie kod.
Zobacz jak będzie różnić się odpowiedź modelu.

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(model="gpt-4o-mini")

limit = 25
prompt = f"Napisz 'Czym jest uczenie maszynowe?' w MAKS {limit} słowach:" # <- podmień tekst
lower_limit = 10
resp = llm.invoke(prompt).content
if len(resp.split()) > lower_limit:
    print("Tekst odpowiedzi jest zbyt długi - wymagane streszczenie. ")
    # szybka korekta – poproś model o skrócenie do limitu
    resp = llm.invoke(f"Skróć to do maks {lower_limit} słów, bez dopisków:\n\n{resp}").content

print(resp)
```
