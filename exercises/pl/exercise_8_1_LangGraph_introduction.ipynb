## Zadania: LangGraph - podstawy


Zadanie 1
State + prosty węzeł
Podmień pytanie przypisane do klucza "question" i uruchom ponizszy kod.

```python
from typing import TypedDict, Optional

class State(TypedDict, total=False):
    question: str
    answer: Optional[str]

def echo_node(state: State) -> State:
    new_state = dict(state)
    if "answer" not in new_state:
        new_state["answer"] = f"ECHO: {state.get('question','')}"
    return new_state

assert callable(echo_node)

state = State({"question": "Czym jest LangGraph?"}) # <- podmień na swój tekst
echo_node(state)
```
