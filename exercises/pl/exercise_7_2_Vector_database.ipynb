Zmodyfikuj poniższy kod we wskazanych miejscach.
Podmień listę tekstów na własne - mogą to być np. skopiowane z dowolnego źródła treści artykułów dotyczacych twoich zainteresowań.
Zmodyfikuj wyszukiwane teksty tak, aby jeden pasował do tematyki jednego z dodanych tekstów, a drugi do innego. Uzyj innych słów w tekstach wyszukiwania niż te zawarte we fragmentach artykułów.

```python
from sentence_transformers import SentenceTransformer
import faiss

model = SentenceTransformer("all-MiniLM-L6-v2")

model.max_seq_length = 256

sentences = [ # zmodyfikuj listę tekstów do dopasowania semantycznego
    "dinosaurs live in africa but in different time dimension",
    "this is sentence about little cat that liked to eat fast food",
    "this is the another sample sentence which is here just to not be matched while other one is"
]

embeddings = model.encode(sentences, normalize_embeddings=True)

d = 384  # dimensions

# Build index
index = faiss.IndexFlatL2(d)  # build the index
index.add(embeddings)  # add vectors to the index

queryText = "french fries" # <- zmodyfikuj wyszukiwany tekst
embeddingSearch = model.encode([queryText], normalize_embeddings=True)
embeddingFound, idx = index.search(embeddingSearch, 1)  # actual search
print(queryText + " matches:\n" + sentences[idx[0][0]])

queryText = "not similar text" # <- zmodyfikuj wyszukiwany tekst
embeddingSearch = model.encode([queryText], normalize_embeddings=True)
embeddingFound, idx = index.search(embeddingSearch, 1)  # actual search
print(queryText + " matches:\n" + sentences[idx[0][0]])
```
